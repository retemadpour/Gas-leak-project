We retrieved estimated population information from American Community Survey site using the following request:<br>
https://api.census.gov/data/2010/acs/acs5/profile?get=group(DP05)&for=tract:*&in=state:36&in=county:*<br>
where DP05 is a group with demographic information.<br>

We make this request for each year 2010-2018. The resulting data contains some punctuation, the column names are codes rather than meaningful words. We need to fix it.


```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
```

<h1><center>DP_05, DEMOGRAPHIC AND HOUSING ESTIMATES </center></h1>

/acs_dp05/data_files/ contains datafiles. In these files names of columns are codes rather than meaningful words.<br>
/acs_dp05/column_names/ contains files with codes explanations for each year. <br>
Same codes have different meaning in different years. 

```python
# create a list of dataframes containing ecoding of the column names
mypath_headers = 'C:/Users/mskac/machineLearning/GasLeakConEd/data/original/acs_data/acs_dp05/column_names/'
head_dfs = []
head_files = [f for f in listdir(mypath_headers) if isfile(join(mypath_headers, f))]

for head_file in head_files:
    head_dfs.append(pd.read_csv(join(mypath_headers, head_file)))
```

```python
print(head_files)
```

```python
#create a list of dataframes with data for each year

dfs = []

mypath_from = 'C:/Users/mskac/machineLearning/GasLeakConEd/data/original/acs_data/acs_dp05/data_files/'
mypath_to = 'C:/Users/mskac/machineLearning/GasLeakConEd/data/processed/acs_data/acs_dp05/'

files = [f for f in listdir(mypath_from) if isfile(join(mypath_from, f))]

for file in files:
    dfs.append(pd.read_csv(join(mypath_from, file)))
```

```python
print(files)
```

<h2> Clean from punctuation </h2>

```python
d = {'[': '', ']': '', '"': ''}
for df in dfs:
    for old, new in d.items():
        df.columns = df.columns.str.replace(old, new)
        for column in df.columns:
            df[column] = df[column].astype(str).str.replace(old, new)
```

```python
dfs[8].head()
```

```python
# there is an extra column in the end, so drop it
for i in range(len(dfs)):
    dfs[i] = dfs[i].drop(dfs[i].columns[-1],axis=1)
```

```python
dfs[1].head()
```

<h2> Rename columns from codes to meaningful names </h2>

```python
# create a list of dictionaries for codes meanings for each year

names = []
for df_head in head_dfs:
    names.append(df_head.to_dict())
```

```python
# for each year compare codes in the dictionary and codes in datafiles. If a code from file is found in a dict - replace with 
# a meaningful name, if not - replace with "unknown"
for index in range (len(dfs)):      
    new_columns = [names[index][column][0] if column in names[index] else "unknown" for column in dfs[index].columns[:-3]]
    new_columns.extend(['state', 'county', 'tract'])
    dfs[index].columns = new_columns
```

```python
for df in dfs:
    df.drop(['unknown'], axis = 1, inplace = True)
```

<h2>Drop unnecessary columns </h2>

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
# columns have different names in different years. If we remove !!Total population from each name, then names will match

for i in range(len(dfs)):
    dfs[i].columns=dfs[i].columns.str.replace('!!Total population','')
```

```python
dfs[3].columns.to_list()
```

```python
# we only want certain columns (new_columns)

final_dfs = []
new_columns = ['Estimate!!SEX AND AGE!!Total population', 
         'Estimate!!SEX AND AGE!!Total population!!Male',
         'Estimate!!SEX AND AGE!!Total population!!Female',
         'Estimate!!SEX AND AGE!!Under 5 years',
         'Estimate!!SEX AND AGE!!5 to 9 years',
         'Estimate!!SEX AND AGE!!10 to 14 years',
         'Estimate!!SEX AND AGE!!Total population!!15 to 19 years',
         'Estimate!!SEX AND AGE!!Total population!!20 to 24 years',
         'Estimate!!SEX AND AGE!!Total population!!25 to 34 years',
         'Estimate!!SEX AND AGE!!Total population!!35 to 44 years',
         'Estimate!!SEX AND AGE!!Total population!!45 to 54 years',
         'Estimate!!SEX AND AGE!!Total population!!55 to 59 years',
         'Estimate!!SEX AND AGE!!Total population!!60 to 64 years',
         'Estimate!!SEX AND AGE!!Total population!!65 to 74 years',
         'Estimate!!SEX AND AGE!!Total population!!75 to 84 years',
         'Estimate!!SEX AND AGE!!Total population!!85 years and over',
         'Estimate!!SEX AND AGE!!Median age (years)',
         'state','county','tract']
for i in range(len(new_columns)):
    new_columns[i] = new_columns[i].replace('!!Total population','')

```

```python
pd.options.mode.chained_assignment = None
for i in range(len(dfs)):
    final_dfs.append(dfs[i][new_columns])
```

```python
# rename columns for easier reffering

for i in range(len(dfs)):
    final_dfs[i] = dfs[i].rename({'Estimate!!SEX AND AGE':'total_population',
                                  'Estimate!!SEX AND AGE!!Male':'male_population',
                                  'Estimate!!SEX AND AGE!!Female':'female_population',
             'Estimate!!SEX AND AGE!!Under 5 years':'population_under5',
             'Estimate!!SEX AND AGE!!5 to 9 years':'population_5_9',
             'Estimate!!SEX AND AGE!!10 to 14 years':'population_10_14',
             'Estimate!!SEX AND AGE!!15 to 19 years':'population_population_15_19',
             'Estimate!!SEX AND AGE!!20 to 24 years':'population_20_24',
             'Estimate!!SEX AND AGE!!25 to 34 years':'population_25_34',
             'Estimate!!SEX AND AGE!!35 to 44 years':'population_35_44',
             'Estimate!!SEX AND AGE!!45 to 54 years':'population_45_54',
             'Estimate!!SEX AND AGE!!55 to 59 years':'population_55_59',
             'Estimate!!SEX AND AGE!!60 to 64 years':'population_60_64',
             'Estimate!!SEX AND AGE!!65 to 74 years':'population_65_74',
             'Estimate!!SEX AND AGE!!75 to 84 years':'population_75_84',
             'Estimate!!SEX AND AGE!!85 years and over':'population_85over',
             'Estimate!!SEX AND AGE!!Median age (years)':'median_age'}, axis = 1)
    final_dfs[i]['population_under10'] = final_dfs[i]['population_under5'] + final_dfs[i]['population_5_9']
    final_dfs[i]['population_10_20'] = final_dfs[i]['population_10_14'] + final_dfs[i]['population_population_15_19']
    final_dfs[i]['population_20_45'] = final_dfs[i]['population_20_24'] + final_dfs[i]['population_25_34'] + final_dfs[i]['population_35_44']
    final_dfs[i]['population_45_65'] = final_dfs[i]['population_45_54'] + final_dfs[i]['population_55_59'] + final_dfs[i]['population_60_64']
    final_dfs[i]['population_65_85'] = final_dfs[i]['population_65_74'] + final_dfs[i]['population_75_84'] 

    final_dfs[i] = final_dfs[i][['total_population','median_age','male_population','female_population','population_under10','population_10_20','population_20_45','population_45_65','population_65_85','population_85over','state','county','tract']]
```

```python
final_dfs[8].head()
```

```python
# find sex ratio (number of men per 1 female)
for i in range(len(final_dfs)):
    final_dfs[i]['sex_ratio'] = 1
    final_dfs[i]['sex_ratio'] = final_dfs[i]['male_population'].astype(int)/final_dfs[i]['female_population'].astype(int)
    final_dfs[i] = final_dfs[i].drop({'male_population', 'female_population'}, axis = 1)
```

```python
# to get geoid we need to make sure state, county and tract columns have lengths of 2, 3, and 6 accordingly, 
# so pad with zeros on the left

for i in range(len(dfs)):
    final_dfs[i]['state'] = final_dfs[i]['state'].str.zfill(2)
    final_dfs[i]['county'] = final_dfs[i]['county'].str.zfill(3)
    final_dfs[i]['tract'] = final_dfs[i]['tract'].str.zfill(6)
    final_dfs[i]['geoid'] = final_dfs[i].state.astype(str)+final_dfs[i].county.astype(str)+final_dfs[i].tract.astype(str)
    final_dfs[i] = final_dfs[i].drop({'state', 'county', 'tract'}, axis = 1)
```

```python
final_dfs[3].head()
```

```python
for index, df in enumerate(final_dfs):
    df.to_csv(join(mypath_to, files[index]))
```

```python
cols_list =[]
for i in range(len(dfs)):
    cols_list.append([c for c in dfs[i].columns if (c[:6] != 'Margin') & (c[:7] != 'Percent') ])
    dfs[i] = dfs[i][cols_list[i]]
```

```python
# columns have different names in different years. If we remove !!Total population from each name, then names will match

for i in range(len(dfs)):
    dfs[i].columns=dfs[i].columns.str.replace('!!Total households','')
    dfs[i].columns=dfs[i].columns.str.replace('Foreign-born population!!','')
    dfs[i].columns=dfs[i].columns.str.replace('Number of grandparents living with own grandchildren under 18 years!!','')
    dfs[i].columns=dfs[i].columns.str.replace('Grandparents responsible for grandchildren', 'Responsible for grandchildren')
    dfs[i].columns=dfs[i].columns.str.replace('Population 25 years and over!!', '')
```

```python
# we only want certain columns (new_columns)

final_dfs = []
new_columns = ['Estimate!!HOUSEHOLDS BY TYPE', 
         'Estimate!!HOUSEHOLDS BY TYPE!!Family households (families)',
         'Estimate!!HOUSEHOLDS BY TYPE!!Nonfamily households',
        'Estimate!!HOUSEHOLDS BY TYPE!!Nonfamily households!!Householder living alone',
 'Estimate!!HOUSEHOLDS BY TYPE!!Nonfamily households!!Householder living alone!!65 years and over',
        'Estimate!!HOUSEHOLDS BY TYPE!!Average household size',
               'Estimate!!HOUSEHOLDS BY TYPE!!Nonfamily households!!Householder living alone!!65 years and over',
               'Estimate!!GRANDPARENTS!!Responsible for grandchildren',
               'Estimate!!DISABILITY STATUS OF THE CIVILIAN NONINSTITUTIONALIZED POPULATION!!Total Civilian Noninstitutionalized Population',
               'Estimate!!U.S. CITIZENSHIP STATUS!!Not a U.S. citizen',
         'state','county','tract']
```

```python
pd.options.mode.chained_assignment = None
for i in range(len(dfs)):
    final_dfs.append(dfs[i][new_columns])
```

```python
final_dfs[0].columns.to_list()
```

```python
# rename columns for easier reffering

for i in range(len(dfs)):
    final_dfs[i] = dfs[i].rename({'Estimate!!HOUSEHOLDS BY TYPE':'total_housholds',
                                  'Estimate!!HOUSEHOLDS BY TYPE!!Family households (families)':'family_housholds',
 'Estimate!!HOUSEHOLDS BY TYPE!!Nonfamily households':'nonfamily_housholds',
 'Estimate!!HOUSEHOLDS BY TYPE!!Nonfamily households!!Householder living alone':'householder_living_alone',
 'Estimate!!HOUSEHOLDS BY TYPE!!Nonfamily households!!Householder living alone!!65 years and over':'householder_over65_living_alone',
 'Estimate!!HOUSEHOLDS BY TYPE!!Average household size':'avg_houshold_size',
 'Estimate!!GRANDPARENTS!!Responsible for grandchildren':'grandparents_responsible_for_grandchildren',
 'Estimate!!DISABILITY STATUS OF THE CIVILIAN NONINSTITUTIONALIZED POPULATION!!Total Civilian Noninstitutionalized Population':'noninstitutionalized_population_with_disability',
 'Estimate!!U.S. CITIZENSHIP STATUS!!Not a U.S. citizen':'not_us_citizen'
                                  }, axis = 1)
    final_dfs[i]['nonfamily_housholds%'] = final_dfs[i]['nonfamily_housholds'].astype(int)/final_dfs[i]['total_housholds'].astype(int)*100
    final_dfs[i]['lonely_housholder%'] = final_dfs[i]['householder_living_alone'].astype(int)/final_dfs[i]['total_housholds'].astype(int)*100
    final_dfs[i]['lonely_housholder_over65%'] = final_dfs[i]['householder_over65_living_alone'].astype(int)/final_dfs[i]['total_housholds'].astype(int)*100
    final_dfs[i]['housholders_grandparents_responsible_for_grandchildren%'] = final_dfs[i]['grandparents_responsible_for_grandchildren'].astype(int)/final_dfs[i]['total_housholds'].astype(int)*100
    final_dfs[i]['population_with_disability(abs)'] = final_dfs[i]['noninstitutionalized_population_with_disability'].astype(int)
    final_dfs[i]['not_us_citizen%'] = final_dfs[i]['not_us_citizen'].astype(int)/final_dfs[i]['total_housholds'].astype(int)*100
    final_dfs[i] =  final_dfs[i][['avg_houshold_size', 'nonfamily_housholds%','lonely_housholder%','lonely_housholder_over65%','housholders_grandparents_responsible_for_grandchildren%','population_with_disability(abs)','not_us_citizen%', 'state', 'county', 'tract']]
```

```python
final_dfs[7].head()
```

```python
# to get geoid we need to make sure state, county and tract columns have lengths of 2, 3, and 6 accordingly, 
# so pad with zeros on the left

for i in range(len(dfs)):
    final_dfs[i]['state'] = final_dfs[i]['state'].astype(int).astype(str).str.zfill(2)
    final_dfs[i]['county'] = final_dfs[i]['county'].astype(int).astype(str).str.zfill(3)
    final_dfs[i]['tract'] = final_dfs[i]['tract'].astype(int).astype(str).str.zfill(6)
    final_dfs[i]['geoid'] = final_dfs[i].state.astype(str)+final_dfs[i].county.astype(str)+final_dfs[i].tract.astype(str)
    final_dfs[i] = final_dfs[i].drop({'state', 'county', 'tract'}, axis = 1)
```
