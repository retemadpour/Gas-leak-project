```python
import pandas as pd
import json
import requests
from urllib.request import urlopen
```

```python
key = '05fcc51d368d71bce9d7627d3097d978'
projectId = '82d5e815'
```

```python
ages = pd.read_csv(r'../data/original/building.csv')
ages.head()
```

```python
ages = ages[['BIN', 'CNSTRCT_YR']]
columns = ['bin', 'yearBuilt']
ages.columns = columns
```

```python
ages['bin']= ages['bin'].astype(int)
```

```python
# api endpoint:
# https://data.cityofnewyork.us/resource/xzdt-eihu.json
# example request: 
# https://api.cityofnewyork.us/geoclient//v1/bin.json?bin=1079043&app_id=abc123&app_key=def456

def getCensusTract(BIN, retryRun=0):  # returns CensusTract
    url = "https://api.cityofnewyork.us/geoclient//v1/bin.json?bin={0}&app_id={1}&app_key={2}".format(BIN, projectId, key)

    if retryRun == 3:  # Failed to get json data 3 times with this longitude and latitude so need to skip this one
        print("failed 3 times")
        return str(0)
    try:
        response = requests.get(url)
        dataJSON = response.json()
        #data = dataJSON["result"]
        #ct = data
        return dataJSON
    except:
        retryRun += 1
        print("******** Error on bin: " + str(BIN) + " ------ retrying " + str(retryRun))
        return getCensusTract(BIN, retryRun)  # *****need to return the recursive function

for row in range(1):
    # b) using the BIN of each entry to find the census data and adding to the respective arrays to add to csv col later
    censusTract = getCensusTract(int(ages.loc[row]["bin"].item()))
                                  # data.iat[i,4] = lat   # data.iat[i,5] = lon
    #df.at[row, "Geoid2010"] = censusTract
    print(censusTract)
```
