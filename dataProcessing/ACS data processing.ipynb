We retrieved estimated population information from American Community Survey site using the following request:<br>
https://api.census.gov/data/2010/acs/acs5/profile?get=group(DP05)&for=tract:*&in=state:36&in=county:*<br>
where DP05 is a group with demographic information.<br>

We make this request for each year 2010-2018. Also, we make requests for different tables: <br>
1. dp02: SELECTED SOCIAL CHARACTERISTICS
2. dp03: DP_03, SELECTED ECONOMIC CHARACTERISTICS
3. dp04: SELECTED HOUSING CHARACTERISTICS
4. dp05: DEMOGRAPHIC AND HOUSING ESTIMATES

The resulting data contains some punctuation, the column names are codes rather than meaningful words. Same codes mean different things for each year. We need to fix all that.
